```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(len(x_train), 28*28) / 255.0
x_test = x_test.reshape(len(x_test), 28*28) / 255.0

def build_model(hidden_layers, neurons, lr):
    model = models.Sequential()
    model.add(layers.Input(shape=(28*28,)))

    for _ in range(hidden_layers):
        model.add(layers.Dense(neurons, activation='relu'))

    model.add(layers.Dense(10, activation='softmax'))

    model.compile(
        optimizer=optimizers.Adam(learning_rate=lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
model_3 = build_model(3, 8, 1e-3)
history_3 = model_3.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_5 = build_model(5, 8, 1e-3)
history_5 = model_5.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_7 = build_model(7, 8, 1e-3)
history_7 = model_7.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_n8 = build_model(3, 8, 1e-3)
history_n8 = model_n8.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_n16 = build_model(3, 16, 1e-3)
history_n16 = model_n16.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_n32 = build_model(3, 32, 1e-3)
history_n32 = model_n32.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_lr2 = build_model(3, 16, 1e-2)
history_lr2 = model_lr2.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```

```python
model_lr4 = build_model(3, 16, 1e-4)
history_lr4 = model_lr4.fit(
    x_train, y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    verbose=0
)
```
